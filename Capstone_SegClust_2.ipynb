### Get and Append Latitude and Longitude Coordinates of Each Neighborhood 

```python
# Install and load Geocoder Library
!pip install msgpack
!pip install geocoder
import geocoder
print('Geocoder installed and imported!')
```

```python
#import time

# initialize your variable to None
lat_lng_coords = None
latitude = []
longitude = []

n=len(pc_table)
postal_code = pc_table['Postcode']

for i in range(n):
    print("row:",i)
    
    # initialize your variable to None
    lat_lng_coords = None
    
    # loop until you get the coordinates
    #start_time = time.time()
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(postal_code[i]))
        lat_lng_coords = g.latlng
    #print("--- %s seconds ---" % round((time.time() - start_time), 2))
    print(g.latlng)

    # Append latitude and longitude lists
    latitude.append(lat_lng_coords[0])
    longitude.append(lat_lng_coords[1])
```

```python
# Append latitude and longitude into dataframe
pc_table["Latitude"] = latitude
pc_table["Longitude"] = longitude

# Show part of revised dataframe
pc_table.head()
```
